```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
# yield of apples (tons per hectare) over six years in an imaginary country called Kanto
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
#Visualizing yield of apples changes over time using line chart
plt.plot(yield_apples)
```

```python
#Customize X axis

years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
plt.plot(years, yield_apples)
```

```python
#Axis Labels

plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

```python
#Plotting multiple lines
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

#Including legend to differentiate between multiple lines
plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges'])
```

```python
#Line Markers
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker= 'x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges'])
```

```python
#Changing the figure size. We us plt.figure() function to change the size of the figure

plt.figure(figsize=(12, 6))
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)")
```

```python
#SCATTER PLOT
#In a scatter plot, the values of 2 variables are plotted as points on a 2-dimensional grid. Additionallly, you can also use a third variable to determine the size or color of the points.

#Iris flowe dataset. Contains sample measurements of sepals and petals for three species of flowers.
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width)
```

```python
#Using scatter plot
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width)
```

```python
#1. We can color the dots using the flower species as a hue
#2. We can also make the points larger using the s argument

#customizing seaborn figures
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100)
```

```python
#2nd approach to plot sns plot using pandas data frame
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', s=100, data=flowers_df)
```

```python
# HISTOGRAM 
# A histogram represents the distribution of a variable by creating interval along the range of values and showing vertical bars to indicvate the number of observations in each bin

plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width)
```

```python
#Control the size of bins or bars
plt.hist(flowers_df.sepal_width, bins=5)
```

```python
#specifying the boundaries of each bin
plt.hist(flowers_df.sepal_width, bins = np.arange(2, 5, 0.25))
```

```python
#MULTIPLE HISTOGRAMS
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
```

```python
#stacking multiple histograms on top of one another
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
         bins=np.arange(2, 5, 0.25),
         stacked=True)

plt.legend(['Setosa', 'Versicolor', 'Virginica'])
```

```python
#BAR PLOT
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
plt.bar(years, oranges)
```

```python
#stack bars on top of each other use bottom argument
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples)
plt.legend(['apples', 'oranges'])
```

```python
#BAR PLOT WITH AVERAGES
#Tips dataset included in seaborn 

tips_df = sns.load_dataset("tips")
tips_df
```

```python
#barplot function can automatically compute avergaes 
#Line in the plot represents variation
sns.barplot(x='day', y='total_bill', data=tips_df)
```

```python
#we can also specify a hue argument to compare bar plots side by side based on a third feature i.e sex
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df)
```

```python
#HEATMAP
#A heatmap is used to visualize two dimensional data like a matrix or a table using colors

flights_df = sns.load_dataset('flights').pivot(index="month", columns="year", values="passengers")
flights_df
```

```python
plt.title("Number of Passengers (1000s)")
sns.heatmap(flights_df)
```

```python
#Display actual value in heat map using annot keyword
plt.title("Number of Passengers (1000s)")
sns.heatmap(flights_df, annot=True, fmt='d', cmap="Blues")
```
